```python
import os
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/garridolecca/AsthmaGIS/master/"
PATH = os.path.join("datasets", "csv")
URL = DOWNLOAD_ROOT + "datasets/csv/Asthma_LA_Enrich_AOI.tar"

def fetch_data(url=URL, path=PATH):
    if not os.path.isdir(path):
        os.makedirs(path)
    tgz_path = os.path.join(path, "Asthma_LA_Enrich_AOI.tar")
    urllib.request.urlretrieve(url, tgz_path)
    tgz = tarfile.open(tgz_path)
    tgz.extractall(path=path)
    tgz.close()
```

```python
fetch_data()
```

```python
import pandas as pd
pd.set_option('display.max_columns',100)

def load_data(path=PATH):
    csv_path = os.path.join(path, "Asthma_LA_Enrich_AOI.csv")
    return pd.read_csv(csv_path)
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(asthma_final, test_size=0.2, random_state=42)
```

```python
train_set.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.1)
save_fig("better_visualization_plot")
```

```python
train_set.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
    s=train_set["Population"]/100, label="Population", figsize=(10,7),
    c="Asthma", cmap=plt.get_cmap("jet"), colorbar=True,
    sharex=False)
plt.legend()
save_fig("housing_prices_scatterplot")
```

```python
import matplotlib.image as mpimg
california_img=mpimg.imread(PROJECT_ROOT_DIR + '/images/end_to_end_project/california.png')
ax = train_set.plot(kind="scatter", x="Longitude", y="Latitude", figsize=(10,7),
                       s=train_set['Population']/100, label="Population",
                       c="Asthma", cmap=plt.get_cmap("jet"),
                       colorbar=False, alpha=0.4,
                      )
plt.imshow(california_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.5,
           cmap=plt.get_cmap("jet"))
plt.ylabel("Latitude", fontsize=14)
plt.xlabel("Longitude", fontsize=14)

prices = train_set["Asthma"]
tick_values = np.linspace(prices.min(), prices.max(), 11)
cbar = plt.colorbar()
cbar.ax.set_yticklabels(["$%dk"%(round(v/1000)) for v in tick_values], fontsize=14)
cbar.set_label('Asthma', fontsize=16)

plt.legend(fontsize=16)
save_fig("california_housing_prices_plot")
plt.show()
```
